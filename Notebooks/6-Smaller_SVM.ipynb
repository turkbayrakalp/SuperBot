```python
import pandas as pd
import numpy as np

df = pd.read_csv('../Dataframes/3-2018_2020Data.csv')
```

```python
df['PercentageDiff'] = ((df['ClosePrice'] - df['OpenPrice']) / df['OpenPrice']) * 100

# Display the updated DataFrame
print(df)
```

```python
# Function to calculate RSI
def calculate_rsi(data, window):
    delta = data.diff()
    gain = ((delta.where(delta > 0, 0)).rolling(window=window).mean())
    loss = ((-delta.where(delta < 0, 0)).rolling(window=window).mean())

    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Function to calculate PSY
def calculate_psy(data, period=6):
    difference = data.diff()
    psy = (difference > 0).rolling(window=period).sum() / period * 100
    return psy

# Function to calculate ROI
def calculate_roi(data, period=1):
    roi = ((data - data.shift(period)) / data.shift(period)) * 100
    return roi

# Add RSI with a 6-day window
df['RSI_6'] = calculate_rsi(df['ClosePrice'], window=6)

# Add PSY with a 6-day window
df['PSY_6'] = calculate_psy(df['ClosePrice'], period=6)

# Add ROI (Return on Investment)
df['ROI'] = calculate_roi(df['ClosePrice'])

# Add Moving Average with a 144-day window
df['MA144'] = df['ClosePrice'].rolling(window=144).mean()

# Add Exponential Moving Average with a 144-day window
df['EMA144'] = df['ClosePrice'].ewm(span=144, adjust=False).mean()

df.head()  # Show the DataFrame head to verify the new columns
```

```python
df = df.dropna()
```

```python
df.describe
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

# Assuming 'df' is your DataFrame

# Convert 'PercentageDiff' to a binary class: 1 for 'buy' (positive) and 0 for 'sell' (negative)
df['Action'] = (df['PercentageDiff'] > 0).astype(int)

# Select features - excluding 'OpenTime' and 'Action'
feature_columns = df.columns.difference(['OpenTime', 'Action'])
X = df[feature_columns]

# Labels
y = df['Action']

# Split the dataset into training (85%) and testing (15%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, shuffle=False)

# Create a pipeline that standardizes, then runs the classifier
# Use LinearSVC with 'l2' penalty for L2 regularization
# To use L1 regularization, set penalty='l1' and loss='squared_hinge' in LinearSVC
# Due to solver restrictions, dual must be False when using 'l1' penalty
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('linear_svc_l2', LinearSVC(penalty='l2', dual=False, max_iter=10000)),  # L2 regularization
    #('linear_svc_l1', LinearSVC(penalty='l1', loss='squared_hinge', dual=False, max_iter=10000))  # L1 regularization
])

# Train the classifier with L2 regularization
pipeline.fit(X_train, y_train)

# Make predictions on the test set
predictions = pipeline.predict(X_test)

# Evaluate the classifier with L2 regularization
print("Evaluation with L2 regularization:")
print(classification_report(y_test, predictions))

# To train and evaluate with L1 regularization, uncomment the 'linear_svc_l1' line and comment out the 'linear_svc_l2' line.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Assuming the rest of the SVM pipeline code has been executed, and you have the predictions and y_test

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, predictions)

# Plot the confusion matrix using Seaborn's heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Sell', 'Buy'], yticklabels=['Sell', 'Buy'])
plt.title('Confusion Matrix')
plt.ylabel('Actual Class')
plt.xlabel('Predicted Class')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
import pandas as pd

# Assuming 'df' is your original DataFrame

# Function to create lagged features for the past candles
def create_lagged_features(df, n_lags=10):
    # Create a new DataFrame to hold the lagged features
    lagged_df = pd.DataFrame(index=df.index)
    # Create lagged features for each of the required columns
    for i in range(1, n_lags + 1):
        lagged_df[f'OpenPrice_lag_{i}'] = df['OpenPrice'].shift(i)
        lagged_df[f'HighPrice_lag_{i}'] = df['HighPrice'].shift(i)
        lagged_df[f'LowPrice_lag_{i}'] = df['LowPrice'].shift(i)
        lagged_df[f'Volume_lag_{i}'] = df['Volume'].shift(i)
        # Additional features can be added here if needed
    # Create the shifted target
    lagged_df['Future_Class'] = df['ClosePrice'].shift(-1) - df['OpenPrice'].shift(-1)
    lagged_df['Future_Class'] = lagged_df['Future_Class'].apply(lambda x: 1 if x > 0 else -1)
    # Drop the NaN values created by the shifts
    lagged_df = lagged_df.dropna()
    return lagged_df

# Use the function to create the lagged features DataFrame
lagged_df = create_lagged_features(df)

# Split into features and target
X = lagged_df.drop(columns=['Future_Class'])
y = lagged_df['Future_Class']

# Split the dataset into training (85%) and testing (15%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, shuffle=False)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the SVM classifier with L2 regularization
svm_classifier = SVC(kernel='rbf')

# Train the classifier
svm_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set
predictions = svm_classifier.predict(X_test_scaled)

# Generate and print a classification report
print(classification_report(y_test, predictions))

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, predictions)
print(conf_matrix)
```
