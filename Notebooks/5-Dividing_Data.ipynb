```python
import pandas as pd
import numpy as np

df = pd.read_csv('../Dataframes/1-Combined_Data.csv')
```

```python
import pandas as pd

# Assuming df is your original DataFrame

# Convert date range to Unix timestamps in milliseconds
start_date = "2018-01-01"
end_date = "2020-01-01"

# Convert to Unix timestamp in milliseconds
start_timestamp_ms = int(pd.Timestamp(start_date).timestamp() * 1000)
end_timestamp_ms = int(pd.Timestamp(end_date).timestamp() * 1000)

# Filter rows where OpenTime is within the specified range, in milliseconds
small_df = df[(df['OpenTime'] >= start_timestamp_ms) & (df['OpenTime'] <= end_timestamp_ms)]
```
