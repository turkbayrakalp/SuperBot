#### In this notebook, the single combined csv file is read and missing time periods are filled with new rows. The resulting csv is saved as `2-Null_Filled_Data.csv`

```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame

# Calculate the time difference between consecutive rows
df['TimeDiff'] = df['OpenTime'].diff()

# Identify where to insert rows (where TimeDiff > 900000)
insert_indices = (df.index[df['TimeDiff'] > 900000]-1).tolist()

new_rows_list = []  # Use a list to collect new rows, for efficiency

# Get the last valid index in your DataFrame
last_valid_index = df.index[-1]

for insert_index in insert_indices:
    # Check if the insert_index is not the last index
    if insert_index != last_valid_index:
        current_time = df.at[insert_index, 'OpenTime']
        
        next_time = df.at[insert_index + 1, 'OpenTime']

        num_inserts = int((next_time - current_time) / 900000) - 1

        for i in range(1, num_inserts + 1):
            new_time = current_time + (i * 900000)
            new_row = [new_time] + [np.nan] * (len(df.columns) - 2)
            new_rows_list.append(new_row)

# Create a DataFrame from the list of new rows
new_rows_df = pd.DataFrame(new_rows_list, columns=df.columns[:-1])  # Exclude TimeDiff from the column names

# Drop the TimeDiff column from the original DataFrame
df.drop(columns=['TimeDiff'], inplace=True)

# Concatenate the original DataFrame with the new rows DataFrame
df = pd.concat([df, new_rows_df], ignore_index=True).sort_values(by='OpenTime').reset_index(drop=True)

print(df)
```
